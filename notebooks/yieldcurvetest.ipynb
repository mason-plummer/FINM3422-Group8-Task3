```python
# Environment and Imports
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

sys.path.append('../src')
from yieldcurve import YieldCurve

print(sys.version)
```

```python
# input yield data 

maturities = [0.25, 0.5, 1.0, 2.0, 5.0, 10.0]
zero_rates = [0.042, 0.043, 0.045, 0.048, 0.052, 0.055]
```

```python
curve = YieldCurve(
    maturities = maturities,
    zero_rates = zero_rates,
    compounding = "continuous"
    )
```

```python
test_maturities = [0.75, 3.0, 7.0]

for T in test_maturities:
    z = curve.get_zero_rate(T)
    d = curve.get_discount_factor(T)
    print(f"T = {T:.2f} years | zero rate = {z:.2%} | discount factor = {d:.4f}")
```

```python
# plot 
curve.plot(max_maturity = 10)
```
